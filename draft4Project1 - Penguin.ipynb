# 3. Building a list of links of the pages of the catalogue
For loop with test code

```python
title2 = []
link2 = []

url_list = [
    'https://www.penguinlibros.com/es/136-libros-feministas?&pageno=1',
    'https://www.penguinlibros.com/es/136-libros-feministas?&pageno=2',
    'https://www.penguinlibros.com/es/136-libros-feministas?&pageno=3',
    'https://www.penguinlibros.com/es/136-libros-feministas?&pageno=4',
    'https://www.penguinlibros.com/es/136-libros-feministas?&pageno=5',
    'https://www.penguinlibros.com/es/136-libros-feministas?&pageno=6',
    'https://www.penguinlibros.com/es/136-libros-feministas?&pageno=7',
    'https://www.penguinlibros.com/es/136-libros-feministas?&pageno=8',
    'https://www.penguinlibros.com/es/136-libros-feministas?&pageno=9']

for url in url_list:
    http = requests.get(url)
    html = http.text
    soup = BeautifulSoup(html)
    book_list = soup.find_all('p',class_='h3 product-title productTitle')
    for a in book_list:
        title2.append(a.text.strip())
        link2.append(a.select("a")[0]["href"])
```

# 5. Looping through the list of books links

- Applying test code

- Listing genres

```python
dates = []
authors = []
genres = []

for link in link2:
    http = requests.get(link)
    html = http.text
    soup = BeautifulSoup(html)
    for date in soup.find('div', class_="product-category-name-editorial text-muted").text.strip().split()[-1:]:
        dates.append(date)
    for author in soup.find('div', class_="product-category-name text-muted py-2 autorYfav"):
        authors.append(author.text.split())
    for genre in soup.find_all('a', class_="contenedorGenerosRelacionados"):
        genres.append(genre.text)


authors
```

```python
authors2 = []
for author in authors:
    author = ''.join(str(author).split(','))
    authors2.append(author)

while "[]" in authors2:
    authors2.remove("[]")
```
